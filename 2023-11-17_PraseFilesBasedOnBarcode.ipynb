```python
import fitz
from pyzbar.pyzbar import decode
from PIL import Image
import re

from os import chdir, rename, remove
from glob import glob as glob


directory = r'C:\Users\hamzat\Desktop\error files'
chdir(directory)

pdf_list = glob('*.pdf')
# print(f'Number of files in directory ' + str(len(pdf_list)))
page_count = 0


def create_jpg_of_page_1(pdf, mat1, mat2):

    doc = fitz.open(pdf)
    # print('in a function')
    # page_count += doc.page_count
    # zoom = 4
    #4,6  only 3 left
    #6,8 - left
    #7,9 - 1 left

    mat = fitz.Matrix(mat1,mat2)
    file_name = doc.name
    # print(file_name)
    count = 0
    image_name = f"{file_name}_page1.jpg"
    # print(doc.pages(0))




    # Count variable is to get the number of pages in the pdf
    for p in doc:
        count += 1
    # print(f'num of pages in this file ' + str(count))
    for i in range(count):
        if i == 0:
            val = f"image_{i+1}.jpg"
            
            page = doc.load_page(0)
            pix = page.get_pixmap(matrix=mat)
            # pix = page.get_pixmap(dpi=1200)
            pix.save(image_name)
            # print('######################')


       
            
    doc.close()
    return image_name

def decode_image(image_name):
    img = Image.open(image_name)
    decoded_list = decode(img)
    img.close()
    # import pdb
    # pdb.set_trace()
    remove(image_name)
    
    

    # print(type(decoded_list))
    
    # print(len(decoded_list))

    Num_of_barcodes = len(decoded_list)
    if Num_of_barcodes == 0:
        return ""
    else: 
        code_data = str(decoded_list[0].data)
        # print(f'code data = {code_data}')
        order_number = re.findall(r'\d+', code_data)
        # print(f'order Number {order_number[0]}')
        return order_number[0]


error_Files = 0
for pdf in pdf_list:
    total_combinitions_tried = 0
    for x, y in [(4, 4),(5, 7), (7,9), (6, 8) ,(4, 6), (2,2),(8,6), (2,4) ]:
        image_name = create_jpg_of_page_1(pdf, x, y)
        order_number = decode_image(image_name)
        total_combinitions_tried += 1
        if total_combinitions_tried == 7:
            error_Files +=1
        if order_number:
            rename(pdf, order_number + '.pdf')
            break
    
    print(f"Total number of combinations tried for {pdf}: {total_combinitions_tried}")
    # print(f'total pages = ' + str(page_count))

print(f'total files ' + str(len(pdf_list)))
print(f'failed  = {error_Files}')




```
